## Improving Plot Aesthetics

### Introduction To The Data

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
women_degrees = pd.read_csv('data/percent-bachelors-degrees-women-usa.csv')
women_degrees.head(15)
```

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'])
plt.show()
```

```python
#dodamo še moške
```

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'], c = 'red', label = 'Women')
plt.plot(women_degrees['Year'], 100-women_degrees['Biology'], c = 'blue', label = 'Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

### Hiding Tick Marks

```python
fig, ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], c = 'red', label = 'Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], c = 'blue', label = 'Men')
ax.tick_params(bottom=False, top=False, right=False, left=False)

plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

### Hiding Spines

```python
print(ax.spines)
```

```python
fig, ax = plt.subplots()
ax.spines
```

```python
fig, ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], label='Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], label='Men')
ax.tick_params(bottom=False, top=False, left=False, right=False)

for key, spine in ax.spines.items():
    spine.set_visible(False)



    
ax.legend(loc='upper right')
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

### Comparing data with multiple plots

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c= 'green', label='Men')
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.tick_params(bottom=False, top = False, left= False, right = False)
    ax.set_title(major_cats[sp])


# Calling pyplot.legend() here will add the legend to the last subplot that was created.
plt.legend(loc='upper right')
plt.show()
```

## Color, Layout, and Annotations

### Color

### Setting Line Color Using RGB

```python
fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    cb_dark_blue = (0/255, 107/255, 164/255) #definicija barve
    cb_orange = (255/255, 128/255,14/255)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(major_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

### Setting Line Width

```python
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1) #z linewidth nastavimo debelino črte
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], linewidth=3, c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], linewidth=2, c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(major_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

### Improve the Layout and Ordering

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1) #z linewidth nastavimo debelino črte
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], linewidth=3, c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], linewidth=1, c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)





plt.legend(loc='upper right')
plt.show()
```

### Replacing the Legend With Annotations

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1) #z linewidth nastavimo debelino črte
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], linewidth=3, c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], linewidth=1, c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    if sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2002, 35, 'Women')

plt.show()
```

## Seaborn

### Introduction to Seaborn

### Introduction to the Data Set

```python
import pandas as pd
titanic = pd.read_csv('data/train.csv')
cols = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
titanic = titanic[cols].dropna() #odvrže stoplce z manjkajočimi vrednostmi

```

```python
titanic.shape
```

```python
titanic.head()
```

### Creating Histograms In Seaborn

```python
import seaborn as sns    # seaborn is commonly imported as `sns` Knjižnica za grafe pri machinelearningu
import matplotlib.pyplot as plt
```

```python
sns.distplot(titanic['Fare']) #dobimo histogram in distribucijsko krivuljo
plt.show()
```

```python
sns.distplot(titanic['Age'])
plt.show()
```

### Generating A Kernel Density Plot

```python
sns.kdeplot(titanic['Age']) #distribucijska krivulja
plt.show()
```

```python
sns.kdeplot(titanic['Age'], shade = True)
plt.xlabel('Age')
plt.show()
```

### Modifying The Appearance Of The Plots

```python
sns.set_style('dark')
sns.kdeplot(titanic['Age'], shade = True)
plt.xlabel('Age')
plt.show()
```

```python
sns.set_style('white')
sns.kdeplot(titanic['Age'], shade = True)
sns.despine(left=True, bottom = True)
plt.xlabel('Age')
plt.show()
```

### Conditional Plots

```python
g = sns.FacetGrid(titanic,col = 'Survived', height =6) #dobimo dva grafa po kategorijah
g.map(sns.kdeplot, 'Age', shade = True)
plt.show()
```

```python
g = sns.FacetGrid(titanic,col = 'Survived', height =6) #dobimo dva grafa po kategoriji preživetja
g.map(plt.hist, 'Age')
plt.show()
```

```python
g = sns.FacetGrid(titanic,col = 'Pclass', height =6) #dobimo tri grafe po kategorijah starosti
g.map(sns.kdeplot, 'Age', shade = True)
plt.show()
```

```python
g = sns.FacetGrid(titanic,col = 'Survived', row ='Pclass') #dobimo 6 grafov po dveh kategorijah
g.map(sns.kdeplot, 'Age', shade = True)
sns.despine(left = True, bottom = True)
plt.show()
```

```python
g = sns.FacetGrid(titanic,col = 'Survived', row ='Pclass', hue ='Sex') #dobimo 6 grafov po treh kategorijah
g.map(sns.kdeplot, 'Age', shade = True)
sns.despine(left = True, bottom = True)
plt.show()
```

###  Adding A Legend

```python
g = sns.FacetGrid(titanic,col = 'Survived', row ='Pclass', hue ='Sex') #dobimo 6 grafov po treh kategorijah
g.map(sns.kdeplot, 'Age', shade = True)
sns.despine(left = True, bottom = True)
g.add_legend()
plt.show()
```
